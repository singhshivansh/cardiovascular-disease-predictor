## Removing the outliers in CVD Data

```python
#removing outliers in diastilic(ap_lo) and systolic(ap_hi) blood pressure
#diastilic blood pressure is always greater than systolic
print("there are {} {:.4f}% cases where diastilic pressure is greater than systolic pressure".format(
    len(data[data['ap_lo'] > data['ap_hi']]), (100)*len(data[data['ap_lo'] > data['ap_hi']])/len(data) ))
```

so there are around difference of 30-40 mm, so we replace the ap_lo value with thier respective values

```python
data[data['ap_lo']>data['ap_hi']].head()
```

```python
#98 percentile and 2 percentile value
ap_lo_v1 = np.percentile(data['ap_lo'],98)
ap_lo_v2 = np.percentile(data['ap_lo'], 2)
print("98 percentile value of diastilic pressure: {}".format(ap_lo_v1))
print("2 percentile value of diastilic pressure: {}".format(ap_lo_v2))
```

```python
ap_hi_v1 = np.percentile(data['ap_hi'], 98)
ap_hi_v2 = np.percentile(data['ap_hi'], 3)

print("98 percentile value of systolic pressure : {}".format(ap_hi_v1))
print("2 percentile value of systolic pressure : {}".format(ap_lo_v2))
```

```python
max(data['ap_lo'])
```

```python
data['ap_lo'][data['ap_lo']>ap_lo_v1] = data['ap_lo'].median()
data['ap_lo'][data['ap_lo'] < ap_lo_v2] = data['ap_lo'].median()
```

```python
print("max:{},  min:{}".format(max(data['ap_lo']), min(data['ap_lo'])))
```

```python
data['ap_hi'][data['ap_hi'] > ap_hi_v1] = data['ap_hi'].median()
data['ap_hi'][data['ap_hi'] < ap_hi_v2] = data['ap_hi'].median()
```

```python
print(max(data['ap_hi']), min(data['ap_lo']))
```

```python
#removing outliers in weigth and height
height_v1 = np.percentile(data['height'], 98)
height_v2 = np.percentile(data['height'], 2)

weight_v1 = np.percentile(data['weight'], 98)
weight_v2 = np.percentile(data['weight'], 2)

for i in ['height', 'weight']:
    print("{} min value:{}  max value:{}".format(i, max(data[i]), min(data[i])))
```

```python
print("the 98 percentile value of height : {}".format(height_v1))
print("the 2 percentile value of height : {}".format(height_v2))

print("the 98 percentile value of weight : {}".format(weight_v1))
print("the 2 percentile value of weight : {}".format(weight_v2))
```

```python
#for height
data['height'][data['height'] > height_v1] = data['height'].median()
data['height'][data['height'] < height_v2] = data['height'].median()

#for weigth
data['weight'][data['weight'] > weight_v1] = data['weight'].median()
data['weight'][data['weight'] < weight_v2] = data['weight'].median()
```
